# Notebook 03: Análisis Comparativo de Perfiles Municipales
En este bloque realizamos un análisis cualitativo y cuantitativo para entender la disparidad en el mercado de Segovia. 
Investigamos por qué ciertos **Centros Comarcales** alejados de la capital presentan niveles de tensión similares o superiores a los **alrededores**, mientras otros municipios de tamaño equivalente se mantienen en la **Zona de Oportunidad**.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Carga de datos
df = pd.read_csv('../data/cleaned/dataset_maestro_segovia.csv', encoding='utf-8-sig', sep=None, engine='python')

# Calculamos el Ratio de Esfuerzo Mensual (Alquiler / (Renta Anual / 12))
df['Esfuerzo_Pct'] = (df['ALQTBID12_M_VC'] / (df['Total'] / 12)) * 100

# Creamos la columna Segmento (repetimos lógica del NB 02 para tenerla disponible)
media_renta = df['Total'].mean()
media_alq = df['ALQTBID12_M_VC'].mean()

def categorizar(row):
    if row['Total'] < media_renta and row['ALQTBID12_M_VC'] > media_alq: return 'Tensionado'
    if row['Total'] > media_renta and row['ALQTBID12_M_VC'] < media_alq: return 'Oportunidad'
    return 'Estándar/Otros'

df['Perfil'] = df.apply(categorizar, axis=1)
print("✅ Datos preparados para comparativa.")
```

## 1. El Esfuerzo Económico: ¿Quién paga más en proporción?
Comparamos los municipios más tensionados frente a los que ofrecen mejores oportunidades para identificar la brecha de accesibilidad.

```python
# Seleccionamos los 5 con más esfuerzo y los 5 con menos esfuerzo
top_esfuerzo = df.nlargest(5, 'Esfuerzo_Pct')
low_esfuerzo = df.nsmallest(5, 'Esfuerzo_Pct')
comparativa_esfuerzo = pd.concat([top_esfuerzo, low_esfuerzo])

plt.figure(figsize=(12, 6))
colors = ['red' if x > df['Esfuerzo_Pct'].mean() else 'green' for x in comparativa_esfuerzo['Esfuerzo_Pct']]
sns.barplot(data=comparativa_esfuerzo, x='NMUN', y='Esfuerzo_Pct', palette=colors)

plt.axhline(df['Esfuerzo_Pct'].mean(), color='blue', linestyle='--', label='Media Provincial')
plt.title('Diferencia de Esfuerzo Económico (%) por Municipio')
plt.xticks(rotation=45)
plt.ylabel('% Renta Mensual dedicada al Alquiler')
plt.legend()
plt.show()
```

## 2. Análisis de Centros Comarcales vs. Alrededores
A continuación, analizamos los municipios grandes que, a pesar de su distancia de la capital, presentan mercados tensionados debido a su autonomía económica e industrial.

```python
# Filtramos municipios de interés (puedes añadir los nombres específicos que comentaste)
# Supongamos que Cuéllar y El Espinar son los grandes tensionados
municipios_interes = df[df['Perfil'].isin(['Tensionado', 'Oportunidad'])].sort_values(by='Esfuerzo_Pct', ascending=False)

# Mostramos solo columnas clave para el análisis
municipios_interes[['NMUN', 'Total', 'ALQTBID12_M_VC', 'Esfuerzo_Pct', 'Perfil']]
```

## 3. Caracterización Numérica de los Segmentos
No basta con ver las barras; necesitamos entender cuánto más caro es, en promedio, vivir en una zona tensionada frente a una de oportunidad. Calculamos las medias de alquiler y esfuerzo por cada perfil detectado.

```python
# Agrupamos por perfil para ver las diferencias reales
resumen_perfiles = df.groupby('Perfil').agg({
    'ALQTBID12_M_VC': 'mean',
    'Total': 'mean',
    'Esfuerzo_Pct': 'mean',
    'NMUN': 'count'
}).rename(columns={'NMUN': 'Número de Municipios'}).round(2)

print("📊 Resumen estadístico por tipo de mercado:")
display(resumen_perfiles)
```

## 4. Variabilidad del Mercado
Utilizamos un diagrama de caja (Boxplot) para observar la dispersión del precio del alquiler en cada segmento. Esto nos permite ver si los municipios "Tensionados" son casos aislados o forman un grupo cohesionado con precios consistentemente altos.

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='Perfil', y='ALQTBID12_M_VC', palette='Set2')
sns.stripplot(data=df, x='Perfil', y='ALQTBID12_M_VC', color='black', alpha=0.3) # Puntos individuales

plt.title('Dispersión de Precios de Alquiler por Perfil de Mercado')
plt.xlabel('Categoría de Municipio')
plt.ylabel('Precio Alquiler Mensual (€)')
plt.show()
```

## 5. Hallazgos Clave del Análisis de Perfiles

Tras el análisis comparativo de los municipios de la provincia de Segovia, se extraen las siguientes conclusiones fundamentalFG:

### 1. Desmitificación de la Distancia (Centros Comarcales)
Se confirma que la tensión del mercado inmobiliario no es exclusi vosa dalrededoresAlfoz de Segovia** (proximidad a la capital). Existen municipios de gran tamaño y alejados geográficamente que presentan niveles de **esfuerzo económico críticos**. 
* Estos núcleos actúan como polos de atracción de empleo local, pero su oferta de vivienda no ha crecido al mismo ritmo que su actividad económica, generando una "tensión por autonomía".

### 2. La Paradoja de los Municipios Similares
El análisis muestra que municipios con demografías y distancias similares a la capital pueden caer en categorías opuestas (**Tensionado vs. Oportunidad**). 
* Esto sugiere que factores cualitativos (presencia de industria específica, políticas de vivienda locales o stock de alquiler disponible) influyen más en el precio final que la simple cercanía a los servicios de la capital.

### 3. Cuantificación de la Brecha de Esfuerzo
* Mientras que en las **Zonas de Oportunidad** el esfuerzo medio se mantiene en niveles saludables (por debajo del 25-30%), en los municipios **Tensionados** la renta dedicada al alquiler supera significativamente la media provincial.
* Esta disparidad pone de manifiesto una segmentación del mercado donde el acceso a la vivienda está condicionado por la ubicación laboral, penalizando a los trabajadores de los centros comarcales más activos.
